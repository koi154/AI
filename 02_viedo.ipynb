## 02. 비디오 입출력
- 비디오의 재생 : 연속된 이미지를 재생하는 것
- 프레임 : 연속된 이미지 중 하나
- 비디오를 재생하기 위해서는 프레임을 연속적으로 실행


```python
import cv2

FIRE_VIDEO_PATH = "videos/fire.mp4"
```

### 2-1. 비디오 파일 입출력
- cv2.VideoCapture("파일경로") : 경로의 파일을 불러옴
- VideoCapture.isOpened() : 파일이 열렸는지 확인
- VideoCapture.read() : 파일을 읽어서 frame을 반환
 - ret : 프레임이 있는지 체크, frame : 불러온 이미지
- "VideoCapture.release() : 영상 객체를 닫기

```python
cap = cv2.VideoCapture(FIRE_VIDEO_PATH)
# print(cap.isOpened()) # 파일이 열렸는지 확인
while cap.isOpened(): #파일이 열려있는동안 반복적으로 실행하겠다.
    ret, frame = cap.read()
    print(ret, frame)

    if not ret:
        break

    cv2.imshow("Video", frame) #비디오라는 창을 띄우고 비디오를 보여준다

    if cv2.waitKey(1) == ord("q"):
        break
    
cap.release()
cv2.destroyAllWindows()


```

### 2-2. 영상의 프레임 불러오기
- 'VideoCapture.get(cv2.CAP_PROP_FPS)' 

```python
cap = cv2.VideoCapture(FIRE_VIDEO_PATH)
fps = cap.get(cv2.CAP_PROP_FPS)
# print(fps)

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break
    
    cv2.imshow("video", frame)

    if cv2.waitKey(int(1000/fps)) == ord("q"): 
        break

cap.release()
cv2.destroyAllWindows()
```

### 2-3. 카메라 입출력


```python
cap = cv2.VideoCapture(0)
# print(cap.isOpened())
if not cap.isOpened():
    exit()

while True:
    ret,frame = cap.read()
    
    if not ret:
        break

    cv2.imshow("Camera", frame)

    if cv2.waitKey(1) == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

### 2-4. 비디오 쓰기
- 'cv2.VideoWriter(파일경로,fourcc,frame,size)' : 영상을 쓰기
- fourcc : 비디오 코덱 정보


```python
cap = cv2.VideoCapture(FIRE_VIDEO_PATH)

# VideiWriter 생성
fourcc = cv2.VideoWriter_fourcc(*"DIVX") # *붙이면 글자가 쪼개짐.
fps = cap.get(cv2.CAP_PROP_FPS)
width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter("output/fire.mp4", fourcc, fps,(width,height)) 

while cap.isOpened():
    ret,frame = cap.read()

    if not ret:
        break

    out.write(frame) # 소리는 들어가지 않음
    cv2.imshow("Video", frame)

    if cv2.waitKey(int(1000/fps)) == ord("q"):
        break

cap.release()
out.release()
cv2.destroyAllWindows()


```
